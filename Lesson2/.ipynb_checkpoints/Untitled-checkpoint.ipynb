```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

y = np.random.randint(10, 101, 50, dtype=int)
x = [x for x in range(0, 50)]

points = {}
for key, value in zip(x,y):
    points[key] = value

for key in range(1, len(x)):
    if points[key] >= points[key-1]:
        del points[key-1]
    else:
        break

del_key = []
for key, value in reversed(list(points.items())):
    
    if points[key] <= points[key-1]:
        del_key.append(key)
    else:
        break

for key in del_key:
    if key in points:
        del points[key]

x = list(points.keys())
y = list(points.values())
plt.plot(x, y, marker='o', linestyle='-')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Графік з точками')

# Показати графік
plt.show()
```

```python
# Знайдемо місця впадин
valleys = []
for i in range(1, len(y) - 1):
    if y[i] < y[i - 1] and y[i] < y[i + 1]:
        valleys.append(i)

# Додамо початок та кінець, якщо вони є впадинами
if y[0] < y[1]:
    valleys.insert(0, 0)
if y[-1] < y[-2]:
    valleys.append(len(y) - 1)

print("Місця впадин:", valleys)
```
