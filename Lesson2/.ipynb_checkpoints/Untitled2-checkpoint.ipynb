```python
import pandas as pd
import numpy as np
```

```python
x = np.array([x for x in range(2, 100)])
y = np.array([y for y in range(2, 100)])
```

```python
r = [[x, y, x * y, x + y] for x in range(2, 100) for y in range(2, 100)]
df = pd.DataFrame(r, columns=['a', 'b', 'dot', 'sum'])
df
```

```python
df_filter = df[df['dot'].map(df['dot'].value_counts()) != 1]
df_filter['dot'].value_counts()
```
