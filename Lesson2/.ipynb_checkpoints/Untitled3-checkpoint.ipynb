```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

y = np.random.randint(10, 101, 50, dtype=int)
x = [x for x in range(0, 50)]

plt.grid(True)
plt.plot(x, y, marker='o', linestyle='-')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Графік з точками')
print(y)
# Показати графік
plt.show()
```

```python
if y[0] > y[1]:
    res_y = [y[0] - 1]
else:
    res_y = [y[0]]

res_y.extend(y)

if y[-1] > y[-2]:
    res_y.append(y[-1] - 1)
else:
    res_y.append(y[-1])

res_x = list(range(len(res_y)))
plt.grid(True)
plt.plot(res_x, res_y, marker='o', linestyle='-')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Графік з точками')
print(res_y)
# Показати графік
plt.show()
```

```python
max_y = []
for key, value in enumerate(res_y):
    if key == 0 or key == len(res_y) - 1:
        continue
    if res_y[key] > res_y[key-1] and res_y[key] > res_y[key+1]:
        max_y.append(key)
max_y
```

```python
result_slice = []

for i in range(len(max_y) - 1):
    result_slice.append([max_y[i], max_y[i+1]])

result_slice
```

```python
coordinats = []
for coordinat in zip(res_x, res_y):
    coordinats.append(coordinat)
coordinats
```

```python
res_coordinats = []
for i_slice in result_slice:
    start_index, end_index = i_slice  # Отримайте початковий та кінцевий індекс
    sliced_coordinats = coordinats[start_index:end_index + 1]  # Витягніть підсписок
    res_coordinats.append(sliced_coordinats)
```

```python
res_coordinats
```

```python
# Створюємо графік
plt.figure(figsize=(10, 6))

# Проходимося по кожному списку в data
for line_number, line_data in enumerate(res_coordinats):
    x_values = [point[0] for point in line_data]
    y_values = [point[1] for point in line_data]
    plt.plot(x_values, y_values, label=f'Line {line_number+1}')

# Налаштовуємо вигляд графіка
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.title('Графік з окремими лініями')

# Показуємо графік
plt.show()
```
