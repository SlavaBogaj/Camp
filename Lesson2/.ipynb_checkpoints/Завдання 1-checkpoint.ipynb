```python
# Імпортуєм бібліотеки
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Створюємо випадковий набір данних
y = np.random.randint(10, 101, 50, dtype=int)
x = [x for x in range(0, 50)]
```

```python
# Створюємо словник
points = {}
for key, value in zip(x,y):
    points[key] = value
```

```python
x = list(points.keys())
y = list(points.values())
plt.plot(x, y, marker='o', linestyle='-')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Графік з точками')

# Показати графік
plt.show()
```

```python
# Перевіряємо края з ліва
for key in range(1, len(x)):
    if points[key] >= points[key-1]:
        del points[key-1]
    else:
        break

# Перевіряємо края з права
del_key = []
for key, value in reversed(list(points.items())):
    
    if points[key] <= points[key-1]:
        del_key.append(key)
    else:
        break

for key in del_key:
    if key in points:
        del points[key]
```

```python
x = list(points.keys())
y = list(points.values())
plt.plot(x, y, marker='o', linestyle='-')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Графік з точками')

# Показати графік
plt.show()
```

```python
# Шукаємо місця впадин
valleys = []
for i in range(1, len(y) - 1):
    if y[i] < y[i - 1] and y[i] < y[i + 1]:
        valleys.append(i)

print("Місця впадин:", valleys)
```

```python
key_list = list(points.keys())
start = key_list[0]
print(start)
end = key_list[-1]
end
```

```python
for i in valleys:
    ls = []
    rs = []
    for j in range(i, start, -1):
        if points[j-1] > points[j]:
            ls.append([j-1, points[j-1]])
        else:
            break
    for s in range(i, end):
        if points[s+1] > points[s]:
            rs.append([s+1])
        else:
            break
    ls = ls[::-1]
    res = ls + [[i, points[i]]] + rs
    print(res)
```
