```python
# Стандартно імпортєм бібліотеки
import pandas as pd
import numpy as np
```

```python
# Створюємо данні з numpy через генератори, можна по різному
x = np.array([x for x in range(2, 100)])
y = np.array([y for y in range(2, 100)])
```

```python
# Створюємо датафрейм викличаємо один з дублікатів 2*3 3*2 і т.д.
filt = []
res = []
for i in x:
    for j in y:
        if j not in filt:
            res.append([i, j, i * j, i +j])
    filt.append(i)
colm = ['a', 'b', 'dot', 'sum']
df = pd.DataFrame(res, columns=colm)
df
```

```python
# Перший фільтр Добуток не знає, це означає що може буди декілька комбінацій
filtered_df_dot_1 = df[df['dot'].map(df['dot'].value_counts()) != 1]
filtered_df_dot_1
```

```python
# Другий фільтр Сумма знає, це означає що може сумма унікальна
filtered_df_sum_1 = filtered_df_dot_1[filtered_df_dot_1['sum'].map(filtered_df_dot_1['sum'].value_counts()) == 1]
filtered_df_sum_1
```

```python
# Третій фільтр
filtered_df_dot_2 = filtered_df_sum_1[filtered_df_sum_1['dot'].map(filtered_df_sum_1['dot'].value_counts()) == 1]
filtered_df_dot_2
```
