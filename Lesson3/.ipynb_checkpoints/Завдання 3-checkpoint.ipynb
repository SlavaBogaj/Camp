### Question 1

```python
import numpy as np
import pandas as pd
```

```python
def answer_one():
    
    url_Energy = './DATA/Energy Indicators.xls'
    energy = pd.read_excel(url_Energy, skiprows=17, skipfooter=38)
    energy.drop(columns=['Unnamed: 0', 'Unnamed: 1'], inplace=True)
    energy.columns=['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
    energy['Country'] = energy['Country'].str.replace(r' \([\w\s]+\)', '', regex=True)
    energy['Country'] = energy['Country'].str.replace(r'\d+', '', regex=True)
    new_name_country = {
    'Republic of Korea': 'South Korea',
    'United States of America': 'United States',
    'United Kingdom of Great Britain and Northern Ireland': 'United Kingdom',
    'China, Hong Kong Special Administrative Region': 'Hong Kong'
    }
    energy['Country'] = energy['Country'].replace(new_name_country)

    energy['Energy Supply'] = energy['Energy Supply'].apply(lambda x: x if isinstance(x, int) else np.nan)
    energy['Energy Supply'] = energy['Energy Supply'] * 1000000
    url_Gdp = './DATA/world_bank.csv'
    GDP = pd.read_csv(url_Gdp, skiprows=4)
    GDP['Country Name'] = GDP['Country Name'].replace({
        "Korea, Rep.": "South Korea",  
        "Iran, Islamic Rep.": "Iran", 
        "Hong Kong SAR, China": "Hong Kong"})
    GDP.rename(columns={'Country Name': 'Country'}, inplace=True)
    url_ScimEn = './DATA/scimagojr-3.xlsx'
    ScimEn = pd.read_excel(url_ScimEn)
    ScimEn = ScimEn.sort_values(by='Rank', ascending=True)
    ScimEn = ScimEn.iloc[:15]
    first = pd.merge(ScimEn, GDP, on='Country', how='inner')
    rez = pd.merge(first, energy, on='Country', how='inner')
    rez = rez[['Country', 'Rank', 'Documents', 'Citable documents', 
               'Citations', 'Self-citations', 'Citations per document', 
               'H index', 'Energy Supply', 'Energy Supply per Capita', 
               '% Renewable', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015']]
    rez.set_index('Country', inplace=True)
    return rez
```

```python
answer_one()
```

### Question 2

```python
def answer_two():
    
    col = np.arange(10,20).tolist()
    dat = answer_one().columns[col]
    avgGDP = answer_one()[dat].mean(axis=1).sort_values(ascending=False)
    
    
    return avgGDP
```

### Question 3

```python
def answer_three():
    
    q4 = answer_two()
    q4 = q4.sort_values(ascending=False)
    r_i = q4.index[5]
    return answer_one().loc[r_i, '2015'] - answer_one().loc[r_i, '2006']
```

### Question 5

```python
def answer_five():
    
    q8 = answer_one()
    q8['pop'] = (q8['Energy Supply'] / q8['Energy Supply per Capita'])
    rez = q8['pop'].sort_values(ascending=False)
    return rez.index[2]
```

```python
def answer_five():

    q9 = answer_one()
    q9['pop'] = (q9['Energy Supply'] / q9['Energy Supply per Capita'])
    q9['Citable documents pop'] = q9['Citable documents'] / q9['pop']
    q9 = q9[['Citable documents pop', 'Energy Supply per Capita']].astype('float64')
    return q9['Citable documents pop'].corr(q9['Energy Supply per Capita'])
```

```python
answer_five()
```

```python
def answer_six():

    q9 = answer_one()
    q9['pop'] = (q9['Energy Supply'] / q9['Energy Supply per Capita'])
    q9['Citable documents pop'] = q9['Citable documents'] / q9['pop']
    q9 = q9[['Citable documents pop', 'Energy Supply per Capita']].astype('float64')
    return q9['Citable documents pop'].corr(q9['Energy Supply per Capita'])
```

### Question 7

```python
def answer_seven():
    ContinentDict = {
        'China': 'Asia',
        'United States': 'North America',
        'Japan': 'Asia',
        'United Kingdom': 'Europe',
        'Russian Federation': 'Europe',
        'Canada': 'North America',
        'Germany': 'Europe',
        'India': 'Asia',
        'France': 'Europe',
        'South Korea': 'Asia',
        'Italy': 'Europe',
        'Spain': 'Europe',
        'Iran': 'Asia',
        'Australia': 'Australia',
        'Brazil': 'South America'
    }

    Top15 = answer_one()
    Top15['pop'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Top15['Continent'] = Top15.index.map(ContinentDict)

    new_df = pd.DataFrame(index=['Asia', 'Australia', 'Europe', 'North America', 'South America'], columns=['size', 'sum', 'mean', 'std'])
    grouped = Top15.groupby('Continent')['pop']

    new_df['size'] = grouped.count()
    new_df['sum'] = grouped.sum()
    new_df['mean'] = grouped.mean()
    new_df['std'] = grouped.std()

    return new_df
```

```python
answer_seven()
```
